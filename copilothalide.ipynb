```python
import halide as hl
import imageio
import numpy as np

input = hl.ImageParam(hl.Float(32), 3)
f = hl.Func('brightened')
x, y, c = hl.Var('x'), hl.Var('y'), hl.Var('c')
f[x, y, c] = hl.min(2.0 * input[x, y, c], 1.0)

img_np = imageio.imread('input.png').astype(np.float32) / 255.0
img_np = np.ascontiguousarray(img_np)
img_np = np.transpose(img_np, (2, 0, 1)).copy()
img = hl.Buffer(img_np)
input.set(img)

output = f.realize([img.width(), img.height(), img.channels()])
output = np.array(output)
output = np.transpose(output, (1, 2, 0))
imageio.imsave('output.png', (output * 255.0).astype(np.uint8))
```

```python
import halide as hl
import imageio
import numpy as np

img_np = imageio.imread("input.png").astype(np.float32) / 255.0
img_np = np.ascontiguousarray(img_np)
img_np = np.transpose(img_np, (2, 0, 1)).copy()
input_buf = hl.Buffer(img_np)

x, y, c = hl.Var("x"), hl.Var("y"), hl.Var("c")
input = hl.ImageParam(hl.Float(32), 3)
input.set(input_buf)

clamped = hl.Func("clamped")
clamped[x, y, c] = input[
    hl.clamp(x, 0, input_buf.width() - 1),
    hl.clamp(y, 0, input_buf.height() - 1),
    c
]

blur_x = hl.Func("blur_x")
blur_x[x, y, c] = (clamped[x-1, y, c] + clamped[x, y, c] + clamped[x+1, y, c]) / 3.0

blur_y = hl.Func("blur_y")
blur_y[x, y, c] = (blur_x[x, y-1, c] + blur_x[x, y, c] + blur_x[x, y+1, c]) / 3.0

output = blur_y.realize([input_buf.width(), input_buf.height(), input_buf.channels()])
output = np.array(output)
output = np.transpose(output, (1, 2, 0))
imageio.imsave("output.png", (output * 255.0).astype(np.uint8))
```

```python
import halide as hl
import imageio
import numpy as np

# Load and convert to grayscale (result will be 2D: H x W)
img_np = imageio.imread("image.png").astype(np.float32) / 255.0
if img_np.ndim == 3:
    img_np = 0.299 * img_np[:, :, 0] + 0.587 * img_np[:, :, 1] + 0.114 * img_np[:, :, 2]
img_np = np.ascontiguousarray(img_np)  # now shape (H, W)

# Create a 2D Halide buffer from the 2D numpy array
input_buf2 = hl.Buffer(img_np)

x, y = hl.Var("x"), hl.Var("y")
input = hl.ImageParam(hl.Float(32), 2)
input.set(input_buf2)

# Use a clamped Func to avoid out-of-bounds indexing at borders
clamped = hl.Func("clamped")
clamped[x, y] = input[
    hl.clamp(x, 0, input_buf2.width() - 1),
    hl.clamp(y, 0, input_buf2.height() - 1),
]

gx = hl.Func("gx")
gy = hl.Func("gy")
gx[x, y] = (-1 * clamped[x-1, y-1] + 1 * clamped[x+1, y-1] +
            -2 * clamped[x-1, y]   + 2 * clamped[x+1, y] +
            -1 * clamped[x-1, y+1] + 1 * clamped[x+1, y+1])
gy[x, y] = (-1 * clamped[x-1, y-1] + -2 * clamped[x, y-1] + -1 * clamped[x+1, y-1] +
             1 * clamped[x-1, y+1] +  2 * clamped[x, y+1] +  1 * clamped[x+1, y+1])

magnitude = hl.Func("magnitude")
magnitude[x, y] = hl.sqrt(gx[x, y]**2 + gy[x, y]**2)

# Realize with the correct 2D dimensions (width, height)
output = magnitude.realize([input_buf2.width(), input_buf2.height()])
output_np = np.array(output)
imageio.imsave("output.png", (np.clip(output_np, 0, 1) * 255).astype(np.uint8))
```

```python
import halide as hl
import imageio
import numpy as np

img_np = imageio.imread("input.png").astype(np.float32) / 255.0
img_np = np.ascontiguousarray(img_np)
img_np = np.transpose(img_np, (2, 0, 1)).copy()
input_buf = hl.Buffer(img_np)

x, y, c = hl.Var("x"), hl.Var("y"), hl.Var("c")
input = hl.ImageParam(hl.Float(32), 3)
input.set(input_buf)

weights = [1, 4, 6, 4, 1]
norm = sum(weights)

clamped = hl.Func("clamped")
clamped[x, y, c] = input[
    hl.clamp(x, 0, input_buf.width()-1),
    hl.clamp(y, 0, input_buf.height()-1),
    c
]

blur_x = hl.Func("blur_x")
blur_x[x, y, c] = sum([weights[i] * clamped[x + i - 2, y, c] for i in range(5)]) / norm

blur_y = hl.Func("blur_y")
blur_y[x, y, c] = sum([weights[i] * blur_x[x, y + i - 2, c] for i in range(5)]) / norm

output = blur_y.realize([input_buf.width(), input_buf.height(), input_buf.channels()])
output_np = np.array(output)
output_np = np.transpose(output_np, (1, 2, 0))
imageio.imsave("output.png", (output_np * 255).astype(np.uint8))
```

```python
import halide as hl
import imageio
import numpy as np

img_np = imageio.imread("input.png").astype(np.float32) / 255.0
img_np = np.ascontiguousarray(img_np)
img_np = np.transpose(img_np, (2, 0, 1)).copy()
input_buf = hl.Buffer(img_np)

x, y, c = hl.Var("x"), hl.Var("y"), hl.Var("c")
input = hl.ImageParam(hl.Float(32), 3)
input.set(input_buf)

clamped = hl.Func("clamped")
clamped[x, y, c] = input[
    hl.clamp(x, 0, input_buf.width()-1),
    hl.clamp(y, 0, input_buf.height()-1),
    c
]

blurred = hl.Func("blurred")
blurred[x, y, c] = (clamped[x-1, y, c] + clamped[x+1, y, c] +
                    clamped[x, y-1, c] + clamped[x, y+1, c] +
                    clamped[x, y, c]) / 5.0

sharpened = hl.Func("sharpened")
sharpened[x, y, c] = hl.clamp(clamped[x, y, c] + (clamped[x, y, c] - blurred[x, y, c]), 0.0, 1.0)

output = sharpened.realize([input_buf.width(), input_buf.height(), input_buf.channels()])
output_np = np.array(output)
output_np = np.transpose(output_np, (1, 2, 0))
imageio.imsave("output.png", (output_np * 255).astype(np.uint8))
```

```python
import halide as hl
import imageio
import numpy as np

img1 = imageio.imread("image.png").astype(np.float32) / 255.0
img2 = imageio.imread("input.png").astype(np.float32) / 255.0
img1 = np.transpose(np.ascontiguousarray(img1), (2, 0, 1)).copy()
img2 = np.transpose(np.ascontiguousarray(img2), (2, 0, 1)).copy()

input1_buf = hl.Buffer(img1)
input2_buf = hl.Buffer(img2)

x, y, c = hl.Var("x"), hl.Var("y"), hl.Var("c")
input1 = hl.ImageParam(hl.Float(32), 3)
input2 = hl.ImageParam(hl.Float(32), 3)
input1.set(input1_buf)
input2.set(input2_buf)

f = hl.Func("added")
f[x, y, c] = hl.clamp(input1[x, y, c] + input2[x, y, c], 0.0, 1.0)

output = f.realize([input1_buf.width(), input1_buf.height(), input1_buf.channels()])
output_np = np.array(output)
output_np = np.transpose(output_np, (1, 2, 0))
imageio.imsave("output.png", (output_np * 255).astype(np.uint8))
```
